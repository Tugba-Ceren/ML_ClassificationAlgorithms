```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x_train=np.array([[160,65],[170,85],[185,85],[188,82],[155,50],[161,58],[180,68],[157,52],[170,66]])
y_train=['K','E','E','E','K','K','E','K','K']
plt.figure(figsize=(12,7))
plt.title('Weight and Height According To Genders')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(Kg)')
for i ,x in enumerate(x_train):
    plt.scatter(x[0] ,x[1], c='k',marker='x' if y_train[i]=='E' else 'D')
plt.grid(True)
plt.show()

```

```python
#Attending numerical value to the gender labels
from sklearn.preprocessing import LabelBinarizer 
lb=LabelBinarizer()
y_train=lb.fit_transform(y_train)
y_train
```
